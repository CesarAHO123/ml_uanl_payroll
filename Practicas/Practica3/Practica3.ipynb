## Escoge de entre los diferentes modelos de regresion (knn, decision tree, svn, logistic regression, )  para crear un modelo con un conjunto de datos de tu eleccion y utiliza cross validation con un criterio a tu eleccion (acccuracy, precision, recall, f1, ROC)  para escoger el mejor modelo para tu criterio.
### El dataset Escogido para la realización de esta practica es el dataset de enfermedades del corazón de el repositorio de UCI. 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import joblib

from ucimlrepo import fetch_ucirepo 
```

```python
heart_disease = fetch_ucirepo(id=45) 
print(heart_disease.variables)
```

### Las variables con las que trabajaremos seran sex, cp, fbs y exang.
#### Sex =Sex (1=male,0=female)
#### cp = Chest Pain (1=typical, 2=atypical,3=non-anginal pain, 4=asymptomatic)
#### fbs= Fasting Blood Sugar (1=true,0=false)
#### exang=exercise induced angina (1=yes,0=no)

```python
X = pd.DataFrame(heart_disease.data.features, columns=['sex', 'cp', 'fbs', 'exang'])
print(X)
```

```python
y = heart_disease.data.targets["num"]
```

```python
# Tratamiento de valores faltantes (eliminación en este caso)
df = pd.concat([X, y], axis=1)
df = df.dropna()
```

```python
selected_features = ['sex', 'cp', 'fbs', 'exang']
```

```python
# Codificación de variables categóricas
df = pd.get_dummies(df, columns=selected_features, drop_first=True)
```

```python
# División del dataset en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Inicialización de los modelos
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
svc = SVC(probability=True)
lr = LogisticRegression(max_iter=1000)
```

```python
# Cross-validation y evaluación de modelos
scoring = 'accuracy' 
knn_scores = cross_val_score(knn, X_train, y_train, cv=5, scoring=scoring)
dt_scores = cross_val_score(dt, X_train, y_train, cv=5, scoring=scoring)
svc_scores = cross_val_score(svc, X_train, y_train, cv=5, scoring=scoring)
lr_scores = cross_val_score(lr, X_train, y_train, cv=5, scoring=scoring)
```

```python
# Resultados
print('KNN Accuracy:', knn_scores.mean())
print('Decision Tree Accuracy:', dt_scores.mean())
print('SVC Accuracy:', svc_scores.mean())
print('Logistic Regression Accuracy:', lr_scores.mean())
```

#### Podemos ver que el mejor modelo es el de Super Vector Classifier

```python
best_model = svc
best_model.fit(X_train, y_train)
```

```python
# Evaluación del mejor modelo en el conjunto de prueba
y_pred = best_model.predict(X_test)
```

```python
print('Test Accuracy:', accuracy_score(y_test, y_pred))
```
